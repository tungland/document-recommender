```python
df = pd.read_csv('/home/lars/app-document-recommender/parliamentary_questions/questions_topics.txt', sep='\t', header=None, index_col=0)
```

```python
df3 = df
```

```python
df2 = df.iloc[:, 1:]
```

# Set input

In the following cell we select the document we want to compare with

```python
selected_doc = df.iloc[300, 1:]
df3['c'] = df2.apply(lambda x : 1 - spatial.distance.cosine(selected_doc, x), axis=1)
```

The following table shows the selected document (cosine similarity of 1.0) and the four most similar results

```python
results = df3.sort_values('c', ascending=False).head()
results.loc[:, [1 ,'c']]
```
